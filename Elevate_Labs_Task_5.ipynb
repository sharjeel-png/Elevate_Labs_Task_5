### Variable Notes

- **pclass**: A proxy for socio-economic status (SES)  
  - 1st = Upper  
  - 2nd = Middle  
  - 3rd = Lower  

- **age**:  
  Age is fractional if less than 1. If the age is estimated, it is in the form of xx.5.

- **sibsp**:  
  The dataset defines family relations in this way:  
  - Sibling = brother, sister, stepbrother, stepsister  
  - Spouse = husband, wife (mistresses and fianc√©s were ignored)

- **parch**:  
  The dataset defines family relations in this way:  
  - Parent = mother, father  
  - Child = daughter, son, stepdaughter, stepson  

  Some children travelled only with a nanny, therefore `parch = 0` for them.

**üîπ CELL 6.1 ‚Äî Outlier Detection using Boxplot (Age)**

```python
# CELL 6.1: Outlier Detection using Boxplot (Age)
# -----------------------------------------------
# This boxplot helps identify outliers and skewness
# in the Age column before any data cleaning.

sns.boxplot(x=df['Age'])

# Add title and label
plt.title('Boxplot of Age (Before Cleaning)')
plt.ylabel('Age')

# Display the plot
plt.show()
```

**# Observation**
- The horizontal boxplot shows several outliers on the higher age side.
- The spread of values is wider towards older ages, indicating right skewness in the Age distribution.
- The presence of these outliers suggests that the mean would be influenced by extreme values.
- Therefore, the median is a more appropriate measure for handling missing Age values.


**üîπ Cell 7: Data Cleaning**

```python
# CELL 7: Data Cleaning (Post-EDA / For Analysis Support)
# ------------------------------------------------------
# We create a cleaned copy of the dataset so that
# the original data remains unchanged.

# Create a copy of the original dataframe
df_clean = df.copy()

# Fill missing Age values using median (robust to outliers)
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())

# Fill missing Embarked values using mode (most frequent category)
df_clean['Embarked'] = df_clean['Embarked'].fillna(df_clean['Embarked'].mode()[0])

# Drop Cabin column due to excessive missing values
df_clean = df_clean.drop(columns=['Cabin'])

# Verify missing values after cleaning
df_clean.isnull().sum()
```

**# Observation**
- Missing values in the Age column were handled using the median, which is appropriate due to the skewed age distribution.
- Missing values in the Embarked column were filled using the mode, as it is a categorical feature.
- The Cabin column was dropped because it contained a large proportion of missing values and would add limited value without extensive feature engineering.
- A cleaned copy of the dataset (`df_clean`) was created to preserve the original dataset (`df`) for reference.

**üîπ CELL 11 ‚Äî Bivariate Analysis: Survival vs Gender**

**üîπ CELL 12 ‚Äî Bivariate Analysis: Survival vs Passenger Class**

**üîπ CELL 13 ‚Äî Univariate Analysis: Fare Distribution**

**üîπ CELL 14 ‚Äî Bivariate Analysis: Survival vs Fare (Boxplot)**

```python
# CELL 14: Bivariate Analysis - Survival vs Fare
# ---------------------------------------------
# This boxplot compares fare distribution for
# passengers who survived vs those who did not.

# Increase figure size for better readability
plt.figure(figsize=(8, 4))

# Create boxplot
sns.boxplot(x='Survived', y='Fare', data=df_clean)

# Add title and labels
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')

# Display the plot
plt.show()
```

**üîπ CELL 15 ‚Äî Multivariate Analysis: Correlation Heatmap**

```python
# CELL 15: Correlation Heatmap
# ---------------------------
# This heatmap shows correlations between numerical variables.

# Increase figure size for readability
plt.figure(figsize=(8, 6))

# Compute correlation matrix for numerical columns
corr_matrix = df_clean.corr(numeric_only=True)

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add title
plt.title('Correlation Heatmap')

# Display the plot
plt.show()
```

**# Observation**
- Survival shows a positive correlation with Fare, indicating that passengers who paid higher fares were more likely to survive.
- Survival has a negative correlation with Pclass, meaning lower passenger class is associated with lower survival.
- Other numerical variables show weak or minimal correlation with survival.
- This suggests that socio-economic factors played a stronger role than family size variables.

**üîπ CELL 16 ‚Äî Multivariate Analysis: Pairplot**

```python
# CELL 16: Multivariate Analysis - Pairplot
# ----------------------------------------
# Pairplot visualizes pairwise relationships between
# numerical variables, colored by survival status.

sns.pairplot(
    df_clean[['Survived', 'Age', 'Fare', 'Pclass']],
    hue='Survived',
    diag_kind='kde'
)

# Display the plot
plt.show()
```

**# Observation**
- The pairplot shows clear separation between survivors and non-survivors based on Fare and Passenger Class.
- Survivors are more concentrated at higher fare values and lower passenger classes.
- Age shows significant overlap between survivors and non-survivors, indicating a weaker relationship with survival.
- These observations are consistent with earlier bivariate and correlation analyses.


**CELL 17 ‚Äî Scatterplot: Age vs Fare**

```python
# CELL 17: Scatterplot - Age vs Fare
# ---------------------------------
# This scatterplot shows the relationship between
# passenger age and fare, colored by survival status.

# Increase figure size for better readability
plt.figure(figsize=(8, 6))

# Create scatterplot
sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Survived',
    data=df_clean
)

# Add title and labels
plt.title('Age vs Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')

# Display the plot
plt.show()
```

**# Observation**
- Passengers who paid higher fares are more likely to have survived across different age groups.
- Survivors are more concentrated in the higher fare range, regardless of age.
- Age alone does not show a strong linear relationship with fare or survival.
- This scatterplot reinforces the influence of socio-economic factors on survival.
